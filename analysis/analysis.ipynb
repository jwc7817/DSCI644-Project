# Quick overall look at missed or other stuff between datasets

This checks for the differences between the datasets to see what changed between them.

```python
import pandas as pd
```

Iowa Liquor Dataset

```python
raw = pd.read_csv("../data/iowa_liquor/iowa_liquor_250.csv")
gt = pd.read_csv("../data/iowa_liquor/mt_cleaned_iowa_liquor_250.csv")
mp = pd.read_csv("../data/iowa_liquor/mp_cleaned_iowa_liquor_250.csv")
llm = pd.read_csv("../data/iowa_liquor/byllm_cleaned_iowa_liquor_250.csv")
```

```python
cols = ['Invoice/Item Number', 'Vendor Number', 'Date', 'Store Number']
print("raw:"); print(raw[cols].head(5).to_string())
print("\nground truth:"); print(gt[cols].head(5).to_string())
print("\nmanual prompting:"); print(mp[cols].head(5).to_string())
print("\nbyLLM:"); print(llm[cols].head(5).to_string())
```

```python
cols_shared = ['Invoice/Item Number', 'Date', 'Store Number', 'Store Name']
print("raw:"); print(raw[cols_shared].head(5).to_string())
print("\nground truth:"); print(gt[cols_shared].head(5).to_string())
print("\nmanual prompting:"); print(mp[cols_shared].head(5).to_string())
print("\nbyLLM:"); print(llm[cols_shared].head(5).to_string())
```
