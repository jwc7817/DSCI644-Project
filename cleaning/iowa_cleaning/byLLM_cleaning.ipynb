## byLLM Script (Iowa)

This is a script based in byLLM to utilize Meaning Typed Programming in order to clean the Iowa dataset. MTP is used to allow the api to interpret what method of cleaning should be done. This utilizes a subset of the 250-row dataset with a slice of 5 columns to account for token limits. The results of this cleaning is compared against our baseline manual-prompting in "DSCI644-Project/analysis". 

The model byLLM is referencing is Google Gemini 3 flash, as it is the latest flash model available for the free tier api. 

### Setup
- Import required libraries 
- Initialize api

```python
import os
import pandas as pd
from byllm.lib import Model, by
import time
from dataclasses import dataclass, asdict

# This pulls the value from the system variable
apiKey = os.getenv('LLM_API_KEY')

if not apiKey:
    raise ValueError("Missing API Key! Please set the LLM_API_KEY environment variable.")

llm = Model(model_name="gemini/gemini-flash-latest" , api_key= apiKey)
```
